```python
def returned_stats (df_transaction, ean):
    
    df_transaction = df_transaction[df_transaction.EAN == ean]
    
    df_resi = df_transaction[df_transaction.QUANTITY < 0]\
        .groupby("COUNTRY").sum().abs().reset_index()\
        .rename(columns={"QUANTITY": "RESI"})
    
    df_venduto = df_transaction[df_transaction.QUANTITY > 0]\
        .groupby("COUNTRY").sum().reset_index()\
        .rename(columns={"QUANTITY": "VENDUTO"})
    
    df_stats_resi = df_resi.merge(df_venduto, on = "COUNTRY", how = 'left')
    df_stats_resi["ratio"] = df_stats_resi["RESI"]/df_stats_resi["VENDUTO"]
        
    
    return df_venduto, df_resi #df_stats_resi.sort_values(by = 'ratio', ascending = False)
```

```python
df_prices = df_prices_read.copy()
to_double_column = list(df_prices.columns.values)
to_double_column.remove('EAN')

df_prices[[*to_double_column]] = df_prices[[*to_double_column]].astype('float')
df_prices_unpivotted = df_prices.melt(id_vars=['EAN'], var_name='COUNTRY', value_name='LIST_PRICE')
df_prices_unpivotted.head(1)
```

#### Data preparation for Transaction

```python
country_adaption = {
    "AUD" : "Australia",
    "BRL" : "Brasil",
    "CAD" : "Canada",
    "CNY" : "China",
    "EUR" : "Undefined_EUR",
    "JPY" : "Japan",
    "MXN" : "Mexico",
    "USD" : "Undefined_USD"
}

df_transactions = df_transaction_read.copy()
df_transactions['COUNTRY'] = df_transactions['COUNTRY'].fillna(df_transactions['CURRENCY'].map(country_adaption))
df_transactions = df_transactions[df_transactions['EAN'].notna()]
df_transactions = df_transactions[df_transactions['QUANTITY'].notna()]
df_transactions["AMOUNT"] = df_transactions["AMOUNT"].astype('double')
df_transactions = df_transactions[df_transactions['AMOUNT'] > 0]
df_transactions["QUANTITY"] = df_transactions["QUANTITY"].astype('int64')
df_transactions["DATE"] = df_transactions["DATE"].str[:10]
df_transactions.loc[df_transactions.COUNTRY == "USA", 'COUNTRY'] = "United States"
df_transactions.tail(1)
```

#### Stats Approach

```python
def for_stats (df_transactions, df_prices_unpivotted):

    df_for_stats = df_transactions[["EAN", "AMOUNT", "QUANTITY", "COUNTRY", "TRANSACTION ID"]]\
                                    .merge(df_prices_unpivotted, on=['EAN', 'COUNTRY'], how='left')
        
    df_for_stats["RATIO_PRICE"] = round(df_for_stats["AMOUNT"] / df_for_stats["LIST_PRICE"],2)
    df_for_stats["DELTA_PRICE"] = df_for_stats["AMOUNT"] - df_for_stats["LIST_PRICE"]
    df_for_stats["TOTAL_AMOUNT"] = df_for_stats["AMOUNT"] * df_for_stats["QUANTITY"]

    df_mean = (pd.DataFrame({'MEAN_PRICE_RATIO' : df_for_stats.groupby(["COUNTRY"])["RATIO_PRICE"]
                .mean().sort_values()})\
                .reset_index())

    ax = df_mean.plot.scatter(x = 'COUNTRY', y = 'MEAN_PRICE_RATIO', grid = True)
    plt.xticks(fontsize=7, rotation=90)
    plt.show()    
    
    return df_for_stats
```

```python
df_for_stats = for_stats (df_transactions, df_prices_unpivotted)
```

### Gli AMOUNT nei casi di South Korea e Giappone sono affetti da un fattore 100 che sembra un errore di valorizzazione nella tabella transactions. Apporto un fattore correttivo a questi casi. Invece, nel caso del Messico, il listino prezzi è errato in media di un fattore 25.

### Per effettuare le stats preferisco apportare dei fattori correttivi.

```python
df_transactions.loc[df_transactions.COUNTRY.isin(["Japan", "South Korea"]), 'AMOUNT'] = df_transactions.AMOUNT/100
df_prices_unpivotted.loc[df_prices_unpivotted.COUNTRY == 'Mexico', 'LIST_PRICE'] = df_prices_unpivotted.LIST_PRICE*25

df_for_stats = for_stats (df_transactions, df_prices_unpivotted)
```
